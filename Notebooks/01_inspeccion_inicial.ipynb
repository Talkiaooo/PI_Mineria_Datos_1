01. Inspección Inicial del Dataset
Este notebook tiene como objetivo realizar una exploración preliminar del dataset provisto por la cátedra. Se busca comprender la estructura de los datos, identificar tipos de variables, detectar la presencia de valores nulos, registros duplicados y evaluar la integridad general de la información antes de proceder con las tareas de limpieza.

```python
import pandas as pd

# Cargar datos
df = pd.read_json('streaming_users_dirty.json')

# 1. Dimensiones
print(f"Dimensiones: {df.shape[0]} filas y {df.shape[1]} columnas.")

# 2. Tipos de datos y nulos
print("\n--- Información general ---")
print(df.info())

# 3. Primeras filas
display(df.head())
```

```python
# 1. Cantidad de nulos por columna
print("--- Valores nulos por columna ---")
print(df.isnull().sum())

# 2. Duplicados
duplicados = df.duplicated().sum()
print(f"\nCantidad de registros duplicados: {duplicados}")
```

```python
# Estadísticas para ver si hay edades raras (ej. 0) o valores extraños
print("--- Estadísticas descriptivas ---")
display(df.describe())
```

Resumen de hallazgos iniciales
Tras la inspección preliminar del dataset, se identificaron cuatro áreas críticas que condicionarán la etapa de preparación:

Inconsistencias categóricas: Las columnas subscription_plan, country y favorite_genre presentan variaciones en la escritura (mayúsculas/minúsculas, abreviaciones y nombres en distintos idiomas), lo cual impedirá una agrupación correcta.

Valores atípicos: Se detectó la presencia de edades registradas como 0 en la columna age, lo cual es técnicamente inconsistente con el servicio.

Integridad de fechas: La columna last_login_date contiene registros con años futuros, sugiriendo errores en la fuente de datos o en la captura.

Completitud: Existen valores faltantes en monthly_watch_time_mins y favorite_genre que deberán ser tratados para evitar sesgos en el análisis multivariado posterior.

Estas observaciones sirven como base fundamental para las acciones de limpieza y normalización que se ejecutarán en la etapa siguiente.

Próximos pasos
